```python
import folium

# Centered on India
m = folium.Map(location=[20.5937, 78.9629], zoom_start=5)
m
```

```python
folium.TileLayer(
    tiles='https://{s}.tile.opentopomap.org/{z}/{x}/{y}.png',
    attr='OpenTopoMap',
    name='Satellite View'
).add_to(m)

folium.LayerControl().add_to(m)

m
```

```python
# Example: Hyderabad region
folium.Marker(
    location=[17.3850, 78.4867],
    popup="Hyderabad",
    tooltip="Click here"
).add_to(m)

m
```

```python
import geopandas as gpd
from shapely.geometry import box

# Define bounding box (Hyderabad area)
bbox = box(78.3, 17.2, 78.6, 17.5)

gdf = gpd.GeoDataFrame({'geometry': [bbox]}, crs="EPSG:4326")
gdf
```

```python
gdf.plot(edgecolor='red', facecolor='none')
```

```python
import folium

# Create base map (India)
m = folium.Map(location=[20.5937, 78.9629], zoom_start=5)

# Hyderabad bounding box coordinates
bounds = [
    [17.2, 78.3],  # bottom-left
    [17.5, 78.6]   # top-right
]

# Add rectangle (bounding box)
folium.Rectangle(
    bounds=bounds,
    color="red",
    fill=True,
    fill_opacity=0.2
).add_to(m)

m
```

```python
folium.Marker(
    location=[17.3850, 78.4867],
    popup="Hyderabad"
).add_to(m)
```
